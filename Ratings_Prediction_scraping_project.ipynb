```python
# scrape all product urls
product_urls = []
start=0
end=18
for page in range(start,end):#for loop for scrapping 18 page
    url=driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')       #scraping urls
    for i in url:
        product_urls.append(i.get_attribute("href"))                        #appending the urls in product_urls list
    nxt_button=driver.find_element_by_xpath('//ul[@class="a-pagination"]/li[4]/a')     #scraping the list of buttons from the page
    nxt_button.click()
    time.sleep(2)
    
    #Make empty list
    Review=[]
    Rating=[]
    
for url in product_urls:
    driver.get(url)
    time.sleep(2)
    
    
    #Extracting Review from xpath
    try:
        review=driver.find_element_by_xpath('//div[@class="a-expander-content reviewText review-text-content a-expander-partial-collapse-content"]')
        Review.append(review.text)
    except NoSuchElementException:
        Review.append("-")
    
    time.sleep(1)
    
    #Extracting Rating from xpath
    try:
        rating=driver.find_element_by_xpath('//span[@class="a-size-base a-nowrap"]/span')
        Rating.append(rating.text)
    except NoSuchElementException:
        Rating.append("-")
        
time.sleep(1)
#Make DataFrame
df=pd.DataFrame({'Review':Review,'Rating':Rating})
df.head(5)        
```
